```python
   from kafka import KafkaProducer
   from datetime import datetime
   import random
   import json
   import time
```

```python
   producer = KafkaProducer(bootstrap_servers='broker:9092', value_serializer=lambda v: json.dumps(v).encode('utf-8'))
   topic_name = 'energy-usage'

   branches = ['Branch1', 'Branch2', 'Branch3', 'Branch4', 'Branch5']

   while True:
       branch = random.choice(branches)
       energy_usage = round(random.uniform(50.0, 500.0), 2)
       data = {
           'branch': branch,
           'energy_usage': energy_usage,
           'timestamp': str(datetime.utcnow())
       }
       producer.send(topic_name, data)
       print(f"Sent data: {data}")
       time.sleep(2)
```
