```python
   from kafka import KafkaConsumer
   import json

   consumer = KafkaConsumer('energy-usage', bootstrap_servers='broker:9092', value_deserializer=lambda m: json.loads(m.decode('utf-8')))

   branch_energy_usage = {}

   for message in consumer:
       data = message.value
       branch = data['branch']
       energy_usage = data['energy_usage']
       timestamp = data['timestamp']

       if branch not in branch_energy_usage:
           branch_energy_usage[branch] = []
       branch_energy_usage[branch].append(energy_usage)

       avg_usage = sum(branch_energy_usage[branch]) / len(branch_energy_usage[branch])

       print(f"Branch: {branch}, Current Energy Usage: {energy_usage} kWh, Average Energy Usage: {avg_usage:.2f} kWh, Timestamp: {timestamp}")

       # Przykładowe wykrywanie anomalii
       if energy_usage > avg_usage * 1.5:
           print(f"ALERT: Anomalous energy usage detected at {branch}!")
```
